```python
from mygrad import Value, MLP, draw_dot

# Test 1: Basic operations
print("Test 1: Basic operations")
a = Value(2.0, label='a')
b = Value(3.0, label='b')
c = a * b + Value(1.0)
c.backward()

graph1 = draw_dot(c)
display(graph1)
print(f"a.grad = {a.grad}, b.grad = {b.grad}") # should be 3.0, 2.0

# Test 2: Single neuron
print("\nTest 2: Simple MLP")
model = MLP(3, [4, 4, 1])
inputs = [2.0, 3.0, -1.0]
output = model(inputs)
print(f"Output: {output.data}")

graph2 = draw_dot(output)
display(graph2)

# Test 3: Backward pass
print("\nTest 3: backward pass")
output.backward()
print(f"Number of parameters: {len(model.parameters())}")
print(f"Sample parameter gradient: {model.parameters()[0].grad}")
graph3 = draw_dot(output)
display(graph3)

print("\ntest concluded")
```
